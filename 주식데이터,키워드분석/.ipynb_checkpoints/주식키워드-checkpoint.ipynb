```python
def clean_issue(issue):
  words = ''
  for line in issue:
    words = words +line
    words = re.sub('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','',words) 
  return words
```

```python
def morph(input_issue) : 
  issue_morphs = okt.morphs(input_issue) 
  stopwords = ['있','지','윈','팩','선','상','유','효','데','부','텍','익','임','로','형', '브','자', '보', '제','콘','티', '엠','계', '알','이','위','쿼','액','전','년','순','량']
  i_list=[word for word in issue_morphs if not word in stopwords]
  return i_list
```

```python
def word_counts(issue_morph):
  count = Counter(issue_morph)
  count.most_common()
  return count.most_common()
```

```python
clean_issues = clean_issue(issue_2019['이슈내용'])
```

```python
morph_issues19= morph(clean_issues)
```

```python
issue19_counts = word_counts(morph_issues19)
```

```python
from gensim.models.word2vec import Word2Vec
```

```python
def word_nouns(input_words):
  input_words = input_words.str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣 ]','',regex=True)
  issue_list = []
  for i in input_words:
    words = okt.nouns(i)
    issue_list.append(words)
  return issue_list
```

```python
def word2Vec(word_llist,word):
  model= Word2Vec(word_llist,window = 10, min_count = 5, workers = 4, sg = 1)
  model.init_sims(replace=True)
  return  model.wv.most_similar(word)
```
