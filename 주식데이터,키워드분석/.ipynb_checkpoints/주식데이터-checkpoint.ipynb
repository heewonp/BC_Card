```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import re
from konlpy.tag import Okt
from konlpy.tag import Kkma
from collections import Counter
import datetime
```

```python
issue = '/content/drive/My Drive/bc카드/이슈지표.csv'
```

```python
issue_data = pd.read_csv(issue, encoding="euc-kr")
issue_data.head()
```

```python
issue_data.info()
```

```python
# words내용만 가져오기
issue_data['이슈내용'] = issue_data['이슈내용'].str.replace('wordweights(.*).','',regex=True)
np.array(issue_data['이슈내용'])
```

```python
issue_data['이슈내용']= issue_data['이슈내용'].str.replace("{'words':",'')
issue_data['이슈내용']
```

```python
issue_data
```

```python
issue_data['주식종목값']= issue_data['주식종목값'].astype(str)
```

```python
words = ''
for line in issue_data['이슈내용'][0:60612]:
    words = words + line
```

```python
 keywords = re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ\\s]','',words)
```

```python
kkma = Kkma()
okt=Okt()
```

```python
test = okt.nouns(keywords)
```

```python
count = Counter(test)
count.most_common()
```
